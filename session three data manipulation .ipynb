## set up enviroment

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
gss.shape
```

## filtering rows
filtering participants who are married

```python
gss_marital=gss.query("marital=='MARRIED'")
gss_marital.head()
```

## filtering rows where age <=20

```python
gss_age = gss.query("age<=20")
gss_age.head()
```

### selecting columns

```python
# columns are selected using loc[]
# selecting "sex" and 'marital'

gss_selected = gss.loc[:, ['sex', 'income','marital','age']]
gss_selected.head()
```

### creating a new variable

```python
# using assign()
gss_mutated=gss.assign(salary_category=np.where(gss['income']>50000,'High', 'Low'))
gss_mutated
```

```python
#sorting data
# The sort_value function
gss_income_arrange=gss.sort_values(by="income", ascending=False)
gss_income_arrange
```

```python
#summarizing data
gss_summarized = gss.groupby('sex')['income'].mean().reset_index()
gss_summarized
```
